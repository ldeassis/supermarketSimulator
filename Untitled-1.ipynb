```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python

# Parâmetros
n = 10000  # Número de ensaios
p = 0.9  # Probabilidade de sucesso

# Gerar amostra Binomial
data = binom.rvs(n, p, size=n)

data = [d/n for d in data]

# Calcular estatísticas
mean = np.mean(data)
p25 = np.percentile(data, 2.5)
p975 = np.percentile(data, 97.5)
erro = p975 - p25
```

```python

# Plotar KDE
sns.kdeplot(data)

# Adicionar linhas verticais para a média e percentis
plt.axvline(mean, color='red', linestyle='dashed', linewidth=0.5)
plt.axvline(p25, color='blue', linestyle='dashed', linewidth=0.5)
plt.axvline(p975, color='blue', linestyle='dashed', linewidth=0.5)

# Adicionar seta entre os percentis
plt.annotate('', xy=(p25, 1), xytext=(p975, 1), arrowprops=dict(arrowstyle='<->'))

# Adicionar legendas acima do gráfico
plt.text(mean, plt.ylim()[1] - 0.5, 'Média', horizontalalignment='center', color='red')
plt.text(p25, plt.ylim()[1] - 0.5, 'P2.5', horizontalalignment='center', color='blue')
plt.text(p975, plt.ylim()[1] - 0.5, 'P97.5', horizontalalignment='center', color='blue')
# Adicionar legenda da margem de erro junto à seta
plt.text((p25+p975)/2, 0.5, 'Margem de erro', horizontalalignment='center')

# Calcular KDE e preencher a parte inferior
kde = sns.kdeplot(data)
x, y = kde.get_lines()[0].get_data()
plt.fill_between(x, y, color="skyblue", alpha=0.5)

# Adicionar rótulo ao eixo x e título ao gráfico
plt.xlabel('Proporção')


# Mostrar gráfico
plt.show()
```
